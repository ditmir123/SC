```python
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    return x

def sigmoid(x):
    return (1/(1+np.exp(-x)))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0,x)

x = np.linspace(-5,5,100)

y_linear = linear(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)

plt.subplot(2,2,1)
plt.plot(x,y_linear)
plt.title('Linear Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(2,2,2)
plt.plot(x,y_linear)
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(2,2,3)
plt.plot(x,y_tanh)
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.subplot(2,2,4)
plt.plot(x,y_relu)
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```
